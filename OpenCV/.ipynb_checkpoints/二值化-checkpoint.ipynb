## 简单阈值 cv2.threshold() 



```python
import cv2
import matplotlib.pyplot as plt
 
img = cv2.imread(r"./images/4.jpg",0) #直接读为灰度图像
# cv2.imshow("img",img)
# cv2.waitKey()
# cv2.destroyAllWindows(0)

ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) # 127是阈值，高于127的就变成白色，低于变成黑色，255是像素值上限
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV) # 127是阈值，高于127的就变成黑色，低于变成白色，255是像素值上限
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC) # 127是阈值，高于127的设为阈值，低于不变，255是像素值上限
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO) # 127是阈值，低于127的设为0，高于不变，255是像素值上限
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV) # 127是阈值，高于127的设为0，低于不变，255是像素值上限
titles = ['img','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img,thresh1,thresh2,thresh3,thresh4,thresh5]
plt.rcParams['figure.dpi'] = 200 #分辨率

# cv2.imshow("v",thresh3)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()

# 这里用plt图像显示有问题 还不知道怎样解决
```

## 自适应阈值 cv2.adaptiveThreshold( )

所有参数：

    第一个：原始图像
    第二个：像素值上限
    第三个：自适应方法Adaptive Method: 
        — cv2.ADAPTIVE_THRESH_MEAN_C ：领域内均值 
        — cv2.ADAPTIVE_THRESH_GAUSSIAN_C ：领域内像素点加权和，权重为一个高斯窗口
    第四个：值的赋值方法：只有cv2.THRESH_BINARY 和cv2.THRESH_BINARY_INV
    第五个:Block size:规定领域大小（一个正方形的领域）
    第六个:常数C，阈值等于均值或者加权值减去这个常数（为0相当于阈值 就是求得领域内均值或者加权值） 
            这种方法理论上得到的效果更好，相当于在动态自适应的调整属于自己像素点的阈值，而不是整幅图像都用一个阈值。

```python
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2) #换行符号 \
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2) #换行符号 \
images = [img,th1,th2,th3]
plt.figure()
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
plt.show()
```
