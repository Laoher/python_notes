```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# opencv之imread()函数

imread(const string& filename, int flag=1)

```python
img = cv.imread(r"./images/pycharm_opencv_logos.jpg",0)
plt.imshow(img,'gray')
```

```python
plt.imshow(~img,'gray') # 这个~的操作可以使所有像素值变成和255的差，也就是说255->0,0->255,240->15,等等
```

# 保存 imwrite

```python
cv.imwrite('r"./images/pycharm_opencv_logos_save.jpg"',img)
```
