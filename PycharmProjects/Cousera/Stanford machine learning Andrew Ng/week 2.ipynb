# Learning Rate

Debugging gradient descent: 画J(θ) 和 梯度下降迭代次数的映射图. 如果 J(θ) 增加了, 就应该减小α.

Automatic convergence test: 如果 J(θ) 减少小于一个阈值就停止,但实际中,这个值很难选

总结:

α 太小: 收敛太慢.

α 太大: 可能不会收敛
