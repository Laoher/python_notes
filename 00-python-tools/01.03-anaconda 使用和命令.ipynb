## conda 命令

### 更新anaconda

conda update conda

conda update anaconda

### 列出所有安装包/查看版本

```python
pip list
```

```python
conda list
```

### 查看具体package的版本

pip list|grep package_name

### 安装，更新，卸载第三方的 **python** 工具包：

conda install <some package>
conda update <some package>
conda remove <some package>

在安装或更新时可以指定安装的版本号，例如需要使用 numpy 1.8.1 ：

conda install numpy=1.8.1
conda update numpy=1.8.1

### 查看 `conda` 的信息：

```python
!conda info
```

# 环境

### 显示环境列表：

conda env list

### 创建新的环境：

conda create -n env_name python=3.4

### 在server上创建环境

conda create --prefix /data/conda/envs/tyler-document python=3.6

### 激活环境

conda activate env_name

### 在server上激活环境

conda activate /data/conda/envs/Tyler_image_classification

### 退出环境

conda deactivate env_name

### 打开anaconda terminal 安装package
